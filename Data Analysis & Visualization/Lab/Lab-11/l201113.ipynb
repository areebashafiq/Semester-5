```python
# import libraries
import plotly as py
import pandas as pd
import numpy as np
```

```python
# creating random data of 1000 rows and 4 columns

df = pd.DataFrame(np.random.randn(1000, 4), columns=['A','B','C','D'])
```

```python
# creating a scatter plot
py.offline.iplot({
    "data": [py.graph_objs.Scatter(x=df['A'], y=df['B'])],
    "layout": py.graph_objs.Layout(title="Scatter Plot")
})
```

```python
# creating a line plot
py.offline.iplot({
    "data": [py.graph_objs.Scatter(x=df.index, y=df['A'])],
    "layout": py.graph_objs.Layout(title="Line Plot")
})
```

```python
# creating a dot plot
py.offline.iplot({
    "data": [py.graph_objs.Scatter(x=df.index, y=df['A'], mode='markers')],
    "layout": py.graph_objs.Layout(title="Dot Plot")
})
```

```python
# creating a pie chart
py.offline.iplot({
    "data": [py.graph_objs.Pie(labels=df.index, values=df['A'])],
    "layout": py.graph_objs.Layout(title="Pie Chart")
})
```

```python
# creating a histogram
py.offline.iplot({
    "data": [py.graph_objs.Histogram(x=df['A'])],
    "layout": py.graph_objs.Layout(title="Histogram")
})
```
