```python
# this file contains an implementation of bi and tri gram

import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('labeled_data.csv')
data=data.head(10).tweet
```

```python
def make_bi_vocab(data):
    unique_words=[]
    for i in range(len(data)):
        temp=data[i].lower().split()
        for j in range(1,len(temp)):
            word= temp[j-1]+' '+temp[j]
            if word not in unique_words:
                unique_words.append(word)
    return unique_words
```

```python
def create_bigram(data):
    vocab=make_bi_vocab(data)
    gram=[]
    for i in range(len(data)):
        out=[]
        temp1={}
        temp=data[i].lower().split()
        for j in range(1,len(temp)):
            word= temp[j-1]+' '+temp[j]
            if word in temp1:
                temp1[word]+=1
            else:
                temp1[word]=1
        for j in vocab:
            if j in temp1:
                out.append(temp1[j])
            else:
                out.append(0)
        gram.append(out)
    return pd.DataFrame(gram,columns=vocab)
```

```python
bi=create_bigram(data)
bi
```

```python
def make_tri_vocab(data):
    unique_words=[]
    for i in range(len(data)):
        temp=data[i].lower().split()
        for j in range(2,len(temp)):
            word= temp[j-2]+' '+temp[j-1]+' '+temp[j]
            if word not in unique_words:
                unique_words.append(word)
    return unique_words
```

```python
def create_trigram(data):
    vocab=make_tri_vocab(data)
    gram=[]
    for i in range(len(data)):
        out=[]
        temp1={}
        temp=data[i].lower().split()
        for j in range(2,len(temp)):
            word= temp[j-2]+' '+temp[j-1]+' '+temp[j]
            if word in temp1:
                temp1[word]+=1
            else:
                temp1[word]=1
        for j in vocab:
            if j in temp1:
                out.append(temp1[j])
            else:
                out.append(0)
        gram.append(out)
    return pd.DataFrame(gram,columns=vocab)
```

```python
tri=create_trigram(data)
tri
```

```python
def create_grams(data,choice):
    if (choice=='bi'):
        return create_bigram(data)
    elif(choice=='tri'):
        return create_trigram(data)
    else:
        print("Invalid choice")
```

```python
create_grams(data,'bi')
```

```python
create_grams(data,'tri')
```
